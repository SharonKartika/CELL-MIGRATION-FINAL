```julia
using Plots, CSV, DataFrames, LaTeXStrings
```

```julia
function getDf(fl="a/catalyst.TXT")
    df = DataFrame(CSV.File("./vidhuldata/$(fl)"))
    dfx = parse.(Float64,string.(df[2:end-1, 1]))
    dfy = parse.(Float64,string.(df[2:end-1, 2]))
    return dfx, dfy 
end
dfx, dfy = getDf()
```

```julia
plot(getDf("a/catalyst.TXT")..., xlims=(460, :auto), ylims=(:auto, 2500), label="PC")
labels = 10:10:60
for i = 1:6
    plot!(getDf("a/a$i.TXT"), label="\$$(labels[i])μL\\;\\;\\;\\mathbf{2} \$  + PC")
end
plot!(xlabel="Wavelength (nm)", ylabel="Emission Intensity", dpi=600)
savefig("vidhuldata/1.png")
savefig("vidhuldata/1.svg")
```

```julia
plot(getDf("b/catalyst b.TXT")..., xlims=(460, :auto), ylims=(:auto, 2500), label="PC")
counti = 1
labelv = 10
for i = 1:6
    plot!(getDf("b/b$i.TXT"), label="\$$(labelv)μL\\;\\;\\;\\textbf{2} \$  + PC")
    counti += 1
    labelv += 10
end
plot!(xlabel="Wavelength (nm)", ylabel="Emission Intensity",dpi=600)
savefig("vidhuldata/2.png")
savefig("vidhuldata/2.svg")
```
